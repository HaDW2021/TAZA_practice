```python
import random
import time


user = input("이름을 입력해 주세요: ")
print(f"{user}님 안녕하세요!") # 위에 있는 유저 네임을 프린트
print("게임 시작! ")
# 게임이라 하긴 뭐하지만, 파이썬 용어를 공부한다는 부담을 줄이기 위한 멘트
# 공부라고 생각하지 말고 게임이라고 생각할 수 있도록

q_list = [
    "abs(", 
    "all(",
    "메서드","메타 문자",
    "문자열 인덱싱","문자열 슬라이싱","__init__","생성자","클래스","객체","lambda","매개변수",
    "패키지","continue","enumerate(","input(","ZeroDivisionError",
    "모듈","변수","인수","import","match(","readline(","result","print(","group("
]



n = 1       # 게임 시도 횟수
cor_cnt = 0 # 정답 개수


start = time.time()  # 왜 중간에 쓰면 에러 날까?

while n <= 5:
    random.shuffle(q_list)
    q = random.choice(q_list)

    print()
    
    print("{}번".format(n))

    print(q) #문제 출력 
    t = input() # 타이핑 입력

    print()

    
    if str(q).strip() == str(t).strip(): #타이핑 t 가 문제 q와 같으면 정답. # 자료에 실수 있을까봐 strip. 안 보이는 공백 때문에 오답 될 수 있으니.
        print("정답!")
        cor_cnt += 1
    else:
        print("...")

    n += 1 # 문제 순번

end = time.time() # 현재시간
et = end - start # 게임 시간
et = format(et, ".0f")

if cor_cnt == 5:
    print("합격")
else :
    print("불합격")

# 수행 시간
print("소요 시간 : ", et, "초", "정답 개수 : {}".format(cor_cnt))

```
